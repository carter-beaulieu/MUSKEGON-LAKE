```python
############################## DO 8m ONLY  #####################################
import sys
import tensorflow as tf
import types

# Create a dummy module for tf.keras.layers.experimental with a preprocessing attribute.
dummy_experimental = types.ModuleType("tensorflow.keras.layers.experimental")
dummy_experimental.preprocessing = tf.keras.layers
sys.modules["tensorflow.keras.layers.experimental"] = dummy_experimental

print("Patched: 'tensorflow.keras.layers.experimental.preprocessing' set to tf.keras.layers.")

# Patch NumPy to support legacy np.object usage.
import numpy as np
np.object = object

import os
import tensorflow.keras.models as km
from autokeras.auto_model import AutoModel
from autokeras.engine.tuner import AutoTuner

import pandas as pd
import matplotlib.pyplot as plt
import logging
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout, Dense, SimpleRNN, LSTM, BatchNormalization
from tensorflow.keras.optimizers import Adam
from sklearn.ensemble import GradientBoostingClassifier
import autokeras as ak

# ----------------------------------------------------------------------------
#  Setup Logging
# ----------------------------------------------------------------------------
logging.basicConfig(format='%(asctime)s - %(levelname)s - %(message)s', level=logging.INFO)
logging.info("Starting Hypoxia Forecasting with Class Weight Adjustments...")

# ----------------------------------------------------------------------------
#  Load and Preprocess Data
# ----------------------------------------------------------------------------
def load_and_preprocess_data(file_path):
    logging.info("Loading dataset from %s", file_path)
    df = pd.read_csv(file_path, parse_dates=["x_date"])
    df.set_index("x_date", inplace=True)
    df.sort_index(inplace=True)
    
    # Create binary target: 1 if do8m ≤ 2 mg/L or 4 mg/L (hypoxia), 0 otherwise.
    df = df.dropna(subset=["do8m"])
    df["label"] = (df["do8m"] <= 4).astype(int)

    #Informed from Permutation Importance and RFE (Recursive Feature Elimination)
    features = ['wt2m', 'wt4m', 'temp_gradient_6m_7m', 'phyco2m', 'ph8m']
    ##['wt2m', 'wt4m', 'temp_gradient_6m_7m', 'phyco2m', 'ph8m']

    #['wt2m', 'wt4m', 'temp_gradient_6m_7m', 'chl2m', 'ph8m']
    #['wt2m', 'wt4m', 'temp_gradient_4m_9m', 'chl2m', 'ph8m']
    

    df = df[features + ["label"]].dropna()  # Label is 1 or 0 for Hypoxia/Not Hypoxia

    Y = df["label"].values
    df.drop(columns=["label"], inplace=True)  # remove target from predictors
    X = df.copy()
    
    scaler = MinMaxScaler()   #Scale feature data between 0 and 1
    X_scaled = scaler.fit_transform(X.values)
    
    window = 14
    x_data, y_data = [], []
    for i in range(window, len(df)):
        x_data.append(X_scaled[i-window:i, :])
        y_data.append(Y[i])
    
    x_data = np.array(x_data)
    y_data = np.array(y_data)

    train_size = int(len(x_data) * 0.75)
    x_train, y_train = x_data[:train_size], y_data[:train_size]
    x_test, y_test = x_data[train_size:], y_data[train_size:]

    # Compute class weights dynamically
    classes = np.unique(y_train)
    class_weights = compute_class_weight("balanced", classes=classes, y=y_train)
    class_weights = {cls: weight for cls, weight in zip(classes, class_weights)}

    logging.info("Training samples: %d, Testing samples: %d", len(y_train), len(y_test))
    logging.info("Computed Class Weights: %s", class_weights)
    
    # Return df as well for plotting purposes
    return x_train, y_train, x_test, y_test, class_weights, df

# ----------------------------------------------------------------------------
# Define Evaluation Functions (Classification)
# ----------------------------------------------------------------------------
def evaluate_classification(y_true, y_prob, model_name, threshold=0.5):
   
    # Convert probabilities to binary predictions using the threshold
    y_pred = (y_prob >= threshold).astype(int)

    # Compute standard classification metrics
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    cm = confusion_matrix(y_true, y_pred)

    # Extract TN, FP, FN, TP from the confusion matrix
    # The confusion matrix for binary classification is in the order:
    # [[TN, FP],
    #  [FN, TP]]
    TN, FP, FN, TP = cm.ravel()

    # Calculate hit rate (True Positive Rate) and false alarm rate (False Positive Rate)
    hit_rate = TP / (TP + FN) if (TP + FN) > 0 else 0
    false_alarm_rate = FP / (FP + TN) if (FP + TN) > 0 else 0

    # Compute Peirce Skill Score (PSS)
    pss = hit_rate - false_alarm_rate

    # Log and print all evaluation metrics
    logging.info("Model: %s | Accuracy: %.3f | Precision: %.3f | Recall: %.3f | F1: %.3f | PSS: %.3f",
                 model_name, acc, prec, rec, f1, pss)
    
    print("\n" + "-"*50)
    print(f"Model: {model_name}")
    print(f"Accuracy           : {acc:.3f}")
    print(f"Precision          : {prec:.3f}")
    print(f"Recall             : {rec:.3f}")
    print(f"F1 Score           : {f1:.3f}")
    print(f"Peirce Skill Score : {pss:.3f}")
    print("Confusion Matrix:\n", cm)
    print("-"*50 + "\n")

    # Return all metrics as a dictionary
    return {"Accuracy": acc, "Precision": prec, "Recall": rec, "F1": f1, "PeirceSkillScore": pss, "ConfusionMatrix": cm}


# ----------------------------------------------------------------------------
# Ensemble Predictions Function (Weighted Average)
# ----------------------------------------------------------------------------
def ensemble_predictions(pred_dict, weights):
    # Ensure each prediction has shape (n_samples, 1)
    first_key = next(iter(pred_dict))
    n_samples = np.array(pred_dict[first_key]).reshape(-1, 1).shape[0]
    ensemble_prob = np.zeros((n_samples, 1))
    total_weight = sum(weights.values())
    for model_name, pred in pred_dict.items():
        pred = np.array(pred).reshape(-1, 1)
        w = weights.get(model_name, 0)
        ensemble_prob += (w / total_weight) * pred
    return ensemble_prob

# ----------------------------------------------------------------------------
# Define Classification Models with Class Weights
# ----------------------------------------------------------------------------
def rnn_model(x_train, y_train, x_test, y_test, class_weights):
    model = Sequential([
        SimpleRNN(100, return_sequences=True, input_shape=(14, x_train.shape[2])),
        Dropout(0.1),
        SimpleRNN(100),
        Dense(1, activation='sigmoid')
    ])
    
    model.compile(optimizer=Adam(0.001), loss='binary_crossentropy', metrics=['accuracy'])
    model.fit(x_train, y_train, batch_size=32, epochs=200, verbose=0, class_weight=class_weights)
    
    return model.predict(x_test)

def lstm_model(x_train, y_train, x_test, y_test, class_weights):
    model = Sequential([
        LSTM(100, return_sequences=True, input_shape=(14, x_train.shape[2])),
        Dropout(0.2),
        LSTM(100),
        Dense(1, activation='sigmoid')
    ])
    
    model.compile(optimizer=Adam(0.001), loss='binary_crossentropy', metrics=['accuracy'])
    model.fit(x_train, y_train, epochs=200, batch_size=32, verbose=0, class_weight=class_weights)
    
    return model.predict(x_test)

def mlp_model(x_train, y_train, x_test, y_test, class_weights):
    x_train_flat = x_train.reshape((x_train.shape[0], -1))
    x_test_flat  = x_test.reshape((x_test.shape[0], -1))

    model = Sequential([
        Dense(128, activation='relu', input_dim=x_train_flat.shape[1]),
        BatchNormalization(),
        Dropout(0.2),
        Dense(64, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    
    model.compile(optimizer=Adam(0.001), loss='binary_crossentropy', metrics=['accuracy'])
    model.fit(x_train_flat, y_train, epochs=300, batch_size=64, verbose=0, class_weight=class_weights)
    
    return model.predict(x_test_flat).reshape(-1, 1)

def gradient_boosting_model(x_train, y_train, x_test, y_test, class_weights):
    x_train_flat = x_train.reshape((x_train.shape[0], -1))
    x_test_flat  = x_test.reshape((x_test.shape[0], -1))

    sample_weights = np.array([class_weights[y] for y in y_train])
    model = GradientBoostingClassifier(n_estimators=100, random_state=42)
    model.fit(x_train_flat, y_train, sample_weight=sample_weights)
    
    return model.predict_proba(x_test_flat)[:, 1].reshape(-1, 1)

# AutoKeras does not support class weights directly
def autokeras_model(x_train, y_train, x_test, y_test):
    x_train_flat = x_train.reshape((x_train.shape[0], -1))
    x_test_flat  = x_test.reshape((x_test.shape[0], -1))
    
    smote = SMOTE(sampling_strategy='auto', random_state=42)
    x_train_resampled, y_train_resampled = smote.fit_resample(x_train_flat, y_train)
    
    clf = ak.StructuredDataClassifier(overwrite=True, max_trials=10)
    clf.fit(x_train_resampled, y_train_resampled, epochs=300, validation_split=0.25, verbose=0)
    
    best_model = clf.export_model()
    return best_model.predict(x_test_flat).reshape(-1, 1)

# ----------------------------------------------------------------------------
# Run Models with Class Weight Adjustments and Evaluate
# ----------------------------------------------------------------------------
input_file = r"C:\Users\cbeau\Downloads\MUSKEGON_LAKE\MLE_2025_SEPERATE\MUSKEGON_LAKE_HYPOXIA_FORECAST_IMPUTED.csv"
x_train, y_train, x_test, y_test, class_weights, df = load_and_preprocess_data(input_file)

# Initialize dictionaries for predictions and evaluation results
predictions = {}
results = {}

model_functions = {
    "RNN": rnn_model, 
    "LSTM": lstm_model, 
    "Gradient Boosting": gradient_boosting_model, 
    "MLP": mlp_model,
    "AutoKeras": autokeras_model
}

for name, model_function in model_functions.items():
    logging.info("Running model: %s", name)
    if name == "AutoKeras":  # AutoKeras does NOT use class weights
        pred = model_function(x_train, y_train, x_test, y_test)
    else:
        pred = model_function(x_train, y_train, x_test, y_test, class_weights)
    
    # Evaluate and store metrics
    metrics = evaluate_classification(y_test, pred, name, threshold=0.5)
    results[name] = metrics
    predictions[name] = pred

# Define ensemble weights (example weights; adjust as needed)
ensemble_weights = {
    "RNN": 0.230,
    "LSTM": 0.208,
    "Gradient Boosting": 0.225,
    "MLP": 0.151,
    "AutoKeras": 0.185
}

#ensemble_weights = {
#     "RNN": 0.194,
#    "LSTM": 0.199,
#    "Gradient Boosting": 0.209,
#    "MLP": 0.192,
#    "AutoKeras": 0.207
#}

ensemble_pred = ensemble_predictions(predictions, ensemble_weights)
ensemble_results = evaluate_classification(y_test, ensemble_pred, "Ensemble (Weighted)", threshold=0.5)
predictions["Ensemble"] = ensemble_pred

print("Finished Training with Class Weight Adjustments")
print("Summary of Evaluation Metrics for All Models:")
for model_name, metrics in results.items():
    print(f"{model_name}: {metrics}")
print(f"Ensemble (Weighted): {ensemble_results}")

# ----------------------------------------------------------------------------
# Plot Forecast Comparisons
# ----------------------------------------------------------------------------
# Set output directory based on input file location.
output_dir = os.path.dirname(input_file)

plt.figure(figsize=(12, 6))
# Build a time index for the test set (accounting for the 14-day window offset)
window = 14
x_data_len = len(df) - window  # valid samples
train_size = int((len(df) - window) * 0.75)
test_index_start = window + train_size
x_dates_test = df.index[test_index_start:]
plt.plot(x_dates_test, y_test, color='black', label='Observed Hypoxia (0/1)', linewidth=1)
plt.plot(x_dates_test, ensemble_pred, color='red', label='Ensemble Predicted Probability')
plt.axhline(0.5, color='gray', linestyle='--', label='Threshold = 0.5')
plt.title("Muskegon Lake Hypoxia Forecast (do8m-based)")
plt.xlabel("Date")
plt.ylabel("Probability of Hypoxia")
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
fig_path = os.path.join(output_dir, "Hypoxia_Forecast_Ensemble_8m_OG.png")
plt.savefig(fig_path, dpi=300)
plt.show()
print(f"Figure saved to: {fig_path}")
```
